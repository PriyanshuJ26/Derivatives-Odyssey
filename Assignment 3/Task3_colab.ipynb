```python
import numpy as np
import matplotlib.pyplot as plt
stock_purchase=100
strike=105
premium=3
S=np.arange(70,131,1)
long_stock=S-stock_purchase
short_call=premium-np.maximum(S-strike,0)
covered_call=long_stock+short_call
breakeven=stock_purchase-premium
max_profit=(strike-stock_purchase)+premium
plt.figure(figsize=(10,6))
plt.plot(S,covered_call,label="Covered Call",linewidth=3)
plt.axhline(0,linestyle="-")
plt.axvline(breakeven,linestyle="-",label=f"Breakeven={breakeven}")
plt.axhline(max_profit,linestyle=":",label=f"Max Profit={max_profit}")
plt.title("Covered Call Payoff Diagram")
plt.xlabel("Stock Price at Expiration")
plt.ylabel("Profit/Loss")
plt.legend()
plt.grid(True)
plt.show()
print("Breakeven =",breakeven)
print("Maximum Profit =",max_profit)
```
